```python
target_name="COPD"
y=df[target_name]
x=df[["Age (in years)","DOE NYHA Class",	"RWMA(seen=1, not seen=0)","Syncope",	"Orthopnoea","Pre-Syncope"]]
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42, shuffle =True)
```

#Models without Optimal Features

1.1) RFC

```python
from sklearn.ensemble import RandomForestClassifier

RFC= RandomForestClassifier()
rf1=RandomForestClassifier(max_samples=0.75,random_state=42)
import datetime
start=datetime.datetime.now()
RFC.fit(x_train,y_train)
rf1.fit(x_train,y_train)
end=datetime.datetime.now()
print("Total time of exection :",end-start)

y_pred1=rf1.predict(x_test)
rfc_pred_proba=rf1.predict_proba(x_test)[:,1]
y_pred=RFC.predict(x_test)
rfc_pred_proba=RFC.predict_proba(x_test)[:,1]

print("Accuracy Score",accuracy_score(y_test,y_pred1)*100)
print("Accuracy Score",accuracy_score(y_test,y_pred)*100)
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Classificaiton Report:\n",classification_report(y_test,y_pred,digits=4))
# print("ROC AUC Score",roc_auc_score(y_test,rfc_pred_proba))
```

1.2) KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

print("Accuracy Score",accuracy_score(y_test,y_pred)*100)
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Classificaiton Report:\n",classification_report(y_test,y_pred,digits=4))

import pickle
filename = 'JSS_model.sav'
pickle.dump(model, open(filename, 'wb'))
load_model = pickle.load(open(filename,'rb'))
load_model.predict([[ 38,0,1,0,0,0]])
```

1.3) Suport Vector Classifier (SVC)

```python
from sklearn.svm import SVC

model=SVC(probability=True)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

print("SVC-Model : Test Accuracy ",round(accuracy_score(y_test,y_pred)*100,2))
print("Classificaiton Report:\n",classification_report(y_test,y_pred,digits=4))
print("SVC-Model : Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
```

1.4) Linear Discriminant Analysis (LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA()
lda.fit(x_train,y_train)

y_pred=lda.predict(x_test)

print("Accuracy Score",round(accuracy_score(y_test,y_pred)*100,2))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Classificaiton Report:\n",classification_report(y_test,y_pred,digits=4))
```
